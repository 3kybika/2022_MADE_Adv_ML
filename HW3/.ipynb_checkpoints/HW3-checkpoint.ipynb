````python
import re
import numpy as np
from tqdm import tqdm
from matplotlib import pyplot as plt
from nltk import ngrams, FreqDist
````

````python
raw_corpus = ""

with open("corpora\AnnaKarenina.txt", "rb") as file:
    raw_corpus = file.read()
    
with open("corpora\WarAndPeace.txt", "rb") as file:
    raw_corpus += file.read()
    
raw_corpus = raw_corpus.decode("utf-8")
raw_corpus[:300]
````

````python
corpus = raw_corpus.lower()
corpus = re.sub(r"[^а-я\s]", "", corpus)
corpus = re.sub(r"\s", " ", corpus)

freq_letters = FreqDist(ngrams(corpus, 1))
freq_letters = {letter[0]: freq_letters[letter] for letter in freq_letters}
````

````python
alphabet = [letter for letter in " абвгдежзийклмнопрстуфхцчшщъыьэюя"]

labels =  [" {} ".format(letter) for letter in alphabet]
labels[0] = "'_'"
labels = np.array(labels)

values = np.array([freq_letters[letter] for letter in alphabet])

sorted_cnt_ind = np.argsort(values)
values = values[sorted_cnt_ind]
labels = labels[sorted_cnt_ind]

x = np.arange(len(labels)) 

plt.figure(figsize=(25, 25))
plt.barh(x, values)
plt.yticks(x, labels, fontsize=15)
plt.xticks(fontsize=15)
plt.title("Letter counts")

for i, v in enumerate(values):
    plt.text(v + 100, i + 0, str(v), color='black', fontweight='bold')

plt.show()
````

Зашифруем кусочки текста методом перестановок:

````python
def gen_rand_permutation_map(alphabet):
    return dict(list(zip(alphabet, np.random.permutation(alphabet))))

def cipher(text, alphabet_mapping):
    return ''.join(alphabet_mapping[c] for c in text)

def get_sorted_ngrams(ngramm_freq):
    cur_freq_cnts = np.array([ngramm_freq[letter] for letter in ngramm_freq])
    cur_freq_ngramms = np.array([''.join(letter) for letter in ngramm_freq])
    cur_freq_idxs = np.argsort(cur_freq_cnts)
    return cur_freq_ngramms[cur_freq_idxs]

def freq_decipher(text, freq_dist, ngramm_len=1):
    cur_freq_dist = FreqDist(ngrams(text, ngramm_len))

    decipher_map = dict(list(zip(
        get_sorted_ngrams(cur_freq_dist),
        get_sorted_ngrams(freq_dist)
    )))
    
    return decipher_map, ''.join([
        decipher_map[''.join(cur_ngram)] if i % ngramm_len == 0 else ''
        for i, cur_ngram in enumerate(ngrams(text, ngramm_len))
    ])

def accuracy(text1, text2):
    right_chars = 0
    for letter_1, letter_2 in zip(text1, text2):
        if letter_1 == letter_2:
            right_chars += 1
    return right_chars / len(text1)
````

````python
test_length = 10_000
start_idx = np.random.randint(0, len(corpus) - test_length)

test_orig = corpus[start_idx:start_idx + test_length]
test_orig[:500]
````

````python
cipher_map = gen_rand_permutation_map(alphabet)
print(cipher_map)
````

````python
ciphered_text = cipher(test_orig, cipher_map)
ciphered_text[:500]
````

````python
decipher_map, deciphered_text = freq_decipher(ciphered_text, freq_letters)
deciphered_text[:500]
````

````python
correct_letters = {}
for letter in alphabet:
    if cipher_map[decipher_map[letter]] == letter:
        correct_letters[letter] = cipher_map[letter]
print(correct_letters)
print("correct letters: {}/{}".format(len(correct_letters.keys()), len(alphabet)))
````

# 3. МСМС-сэмплирование 
Но и это ещё не всё: биграммы скорее всего тоже далеко не всегда работают. Основная часть задания — в том, как можно их улучшить:
- предложите метод обучения перестановки символов в этом задании, основанный на MCMC-сэмплировании, но по-прежнему работающий на основе статистики биграмм;
- реализуйте и протестируйте его, убедитесь, что результаты улучшились.

````python
def get_freq_map(text, ngramm_len=2):
    freq_bigrams = FreqDist(ngrams(text, ngramm_len))
    len_text = len(text)
    ngramm_cnt = len(set(text))**ngramm_len
    
    return {''.join(letter): freq_bigrams[letter] / (len_text + ngramm_cnt) for letter in freq_bigrams}
    

def get_text_likelihood(text, mapping, bigram_probs, ngramm_len=2):
    res_proba = 0
    decoded_text = ''.join([mapping[letter] for letter in text])
    for i in range(len(decoded_text) - ngramm_len):
        cur_bigram = decoded_text[i: i + ngramm_len]
        bigram_proba = bigram_probs.get(cur_bigram)
        
        if bigram_proba is None:
            bigram_proba = 1 / (len(text) + len(alphabet)**ngramm_len)
            
        res_proba += np.log(bigram_proba)
        #res_proba += bigram_proba
    return res_proba


def mcmc_decipher(text, target_alphabet, source_alphabet, freq_map, trials_num=10, iters_per_trial=1000, ngramm_len=2):
    best_likelihood_mapping = None
    best_likelihood = 0
    cur_trial_alphabet = list(source_alphabet)

    for _ in tqdm(range(trials_num), leave=False, position=0, total=trials_num):
        
        cur_trial_alphabet = list(source_alphabet)
        cur_mapping = dict(zip(target_alphabet, cur_trial_alphabet))
        cur_likelihood = get_text_likelihood(text, cur_mapping, freq_map, ngramm_len)

        for _ in range(iters_per_trial):

            for _ in range(1):
                new_alphabet = cur_trial_alphabet.copy()
                swap_1, swap_2 = np.random.choice(len(new_alphabet), replace=False, size=2)
                new_alphabet[swap_1], new_alphabet[swap_2] = new_alphabet[swap_2], new_alphabet[swap_1]
    
            new_mapping = dict(zip(target_alphabet, new_alphabet[:len(target_alphabet)]))
            new_likelihood = get_text_likelihood(text, new_mapping, freq_map, ngramm_len)

            if  new_likelihood - cur_likelihood >  0:
                cur_trial_alphabet = new_alphabet
                cur_likelihood = new_likelihood
                cur_mapping = new_mapping

        if cur_likelihood > best_likelihood or best_likelihood_mapping is None:
            best_likelihood = cur_likelihood
            best_likelihood_mapping = cur_mapping

    print(f'Best likelihood: {best_likelihood}')
    decipher_map = best_likelihood_mapping
    print(decipher_map)
    
    return decipher_map, ''.join([
        decipher_map[letter] for letter in text
    ])
````
